```
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

__Učitavamo podatke iz izveštaja o svetskoj sreći iz 2019. godine__

Kolone:
Overall rank - rang u svetu

Country or region - ime države ili regiona

Score - koeficijent sreće

GDP per capita - Rast gdpa

Social support - zadvoljstvo sa prijateljima i porodicom

Healthy life expectancy - zdravlje i energija za svakodnevne aktivnosti

Freedom to make life choices - mogućnost izbora

Generosity - Darežljivost

Perceptions of corruption - korupcija

Izvor:https://www.kaggle.com/datasets/priyanka841/2019-world-happiness-report-csv-file

```
file_location = "/FileStore/tables/2019.csv"
hapDF = spark.read.option("inferSchema", "true").option("header", "true").csv(file_location)
hapDF.display()
hapDF.printSchema()
```

__Preimenujemo kolone__

```


columns = ["rank","country","score","gdp_growth","social_support","health","freedom","generosity","corruption"]
old = ["Overall rank", "Country or region", "Score", "GDP per capita", "Social support", "Healthy life expectancy", "Freedom to make life choices", "Generosity", "Perceptions of corruption"]
num = 0
for c in columns:
    hapDF = hapDF.withColumnRenamed(old[num],c)
    num += 1
hapDF.display()
```

__Stavimo da tekst u koloni bude ispisan malim slovima__

```
hapDF = hapDF.withColumn("country",lower(col("country")))
hapDF.display()
```

__Promenimo vrednosti kolona da se poklapaju sa vrednostima u drugoj tabeli__

```
hapDF = hapDF.withColumn("country", when(hapDF.country == "congo (kinshasa)", "congo, dem. rep.")\
                            .when(hapDF.country == "congo (brazzaville)", "congo, rep.")\
                            .otherwise(hapDF.country))
```

__Izbacujemo ne korišćene kolone__

```
hapDF = hapDF.drop("gdp_growth", "social_support", "health", "freedom", "generosity", "corruption")
display(hapDF)
```

__Ućitamo podatke iz izveštaja o stopi samoubistva i ubistva od 1970-2020 godine__

Kolone:

country - ime države

iso3c - standard za skraćenice država

iso2c - Američki standard za skraćenice država

year - godina

intentional homicides (per 100,000 people) - namerna ubistva na 100,000 čoveka

Suicide mortality rate (per 100,000 population)

GDP (current US$) - gdp u toj godini

GDP per capita, PPP(current international $) - PPP u toj godini

adminregion - administrativna regija

incomeLevel - nivo prihoda

Izvor:https://www.kaggle.com/datasets/prasertk/homicide-suicide-rate-and-gdp

```
file_location = "/FileStore/tables/global_suicide.csv"
suDF = spark.read.option("inferSchema", "true").option("header", "true").csv(file_location)
suDF.display()
suDF.printSchema()
```

__Preimenujemo kolonu radi lakšeg rukovanja kasnije__

```
suDF = suDF.withColumnRenamed('Suicide mortality rate (per 100,000 population)', 'suicide_rate').withColumnRenamed('GDP (current US$)', 'gdp')

suDF.display()
```

__Uključimo samo podatke iz 2019. godine__

```
suDF = suDF.where(suDF.year == 2019)

suDF.display()
```

__Izbacimo nepotrebne kolone__

```
suDF = suDF.drop('iso2c', 'Intentional homicides (per 100,000 people)', 'GDP per capita, PPP (current international $)', 'adminregion', 'incomeLevel')
suDF.display()
```

__Izbacimo redove koji imaju null vrednost__

```
suDF = suDF.where((suDF.suicide_rate.isNotNull()) & (suDF.iso3c.isNotNull()) & (suDF.gdp.isNotNull()))

suDF.display()
```

__Stavimo da sva slova budu mala, da bi osigurali da sve države budu isto napisane__

```
suDF = suDF.withColumn("country",lower(col("country")))
suDF.display()
```

__Promenimo vrednosti kolone kako bi se poklapale sa vrednostima u drugoj tabeli__

```
suDF = suDF.withColumn("country", when(suDF.country == "russian federation", "russia")\
                            .when(suDF.country == "slovak republic", "slovakia")\
                            .when(suDF.country == "iran, islamic rep.", "iran")\
                            .when(suDF.country == "egypt, arab rep.", "egypt")
                            .otherwise(suDF.country))
```
