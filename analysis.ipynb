```
#Uƒçitamo transformisane podatke
file_path = "/tmp/alcohol/"

alcDF = spark.read.csv(file_path, header=True, inferSchema=True)
alcDF.display()
```

__Prikaz mape sveta sa stopom samoubistva__

```
#Uƒçitamo transformisane podatke
file_path = "/tmp/suicide/"

suDF = spark.read.csv(file_path, header=True, inferSchema=True)
suDF.display() 
```

__Uticaj alkohola na sreƒáu naroda u dr≈æavi__

```
joinedDF = newHapDF.join(newAlcDF, newHapDF.country == newAlcDF.country, "inner").drop(newAlcDF.country).drop(newAlcDF.gdp).select("rank","country","score","both","male","female", "gdp")
joinedDF.display()
df = joinedDF.sort(joinedDF.score.desc())
```

```

df.createOrReplaceTempView("df_table")
print(df.head())
print(df.tail(1))
```

Iz gornjih podataka mo≈æemo videti da Finska ima najveci koeficijent sreƒáe 7.769, a Centralnoafriƒçka republika ima najmanji koeficijent sreƒáe 3.083. Takoƒëe mo≈æemo primetiti da Finska konzumira 10.65 litra ƒçistog alkohola 

godi≈°nje, ≈°to je znazno veƒáe od Centralnoafriƒçke republike koja konzumira 1.71 litra ƒçistog alkogola godi≈°nje. Ovo mo≈æe da znaƒçi da je Finska sreƒánija jer konzumira vi≈°e alkohola, ali mo≈æe biti i do razlike u ekonomskoj

razvijenosti zemlje. Za sledeƒái primer uzeƒáemo zemlje sa sliƒçnim ekonomskim stanjem.

```
%sql
SELECT country, score, both, gdp from df_table where country == "france" OR country == "italy"
```

U gornjem primeru vidimo da Francuska konzumira vi≈°e alkohola i da je koeficijent sreƒáe veƒái u odnosu na Italiju

```
%sql
SELECT country, score, both, gdp from df_table where country == "germany" OR country == "austria"
```

U gornjem primeru vidimo da iako Nemaƒçka konzumira vi≈°e alkohola, nema veƒái koeficijent sreƒáe.
```
%sql
SELECT country, score, both, gdp from df_table where country == "latvia" OR country == "turkey"
```

Za poslednji primer ƒáemo uzeti Latviu i Tursku. Latvija je zemlja sa najvi≈°om konzumpcijom alkohola, a Turska je zemlja sa veoma niskom konzumpcijom alkohola. Razlika u konzumpciji alkohola je ogromna u odnosu na razliku u

__Da li sreƒánije zemlje nu≈æno imaju manju stopu samoubistava__

```
joinedDF = hapDF.join(suDF, hapDF.country == suDF.country, "inner").drop(suDF.country).select("rank", "country","score","suicide_rate", "gdp")
display(joinedDF)
```

U gornjem primeru mo≈æemo da vidimo da zemlje sa vi≈°im koeficijentom sreƒáe mogu imati skoro duplo veƒáu stopu samoubistva od zemlja sa ni≈æim koeficijentom sreƒáe

```
joinedDF = joinedDF.sort(joinedDF.score.desc())
df = joinedDF
df.createOrReplaceTempView("df_table")
```

```
display(df.take(10))
display(df.tail(10))

```

U gornjem primeru imamo 10 zemlja sa najveƒáim koeficijentom sreƒáe i 10 zemlja sa najni≈æim koeficijentoim sreƒáe. Mo≈æemo primetiti da FInska koja ima najveƒái koeficijent sreƒáe ima skoro 4 puta veƒáu stopu samoubistva od Afganistana koji ima duplo manji koeficijent sreƒáe

```
df = df.sort(df.suicide_rate.desc())
display(df.take(10))
display(df.tail(10))
```

U gornjem primeru imamo 10 zemlja sa najveƒáom stopom samoubistva i 10 zemlja sa najni≈æom stopom samoubistva. U prvom grafu se veƒáinski nalaze zemlje kategarizovane u zemlje drugog sveta, pored Lesata koji je

zemlja treƒáeg sveta i Slovenije koja mo≈æe da se smatra zemljom prvog sveta. U donjem grafu su sve zemlje treƒáeg sveta. Zdravorazumski zar ne bi trebalo da razvijenije zemlje imaju veƒái koeficijent sreƒáe i manju stopu samoubistva

ali ovde vidimo da to nije uvek istina.

__Rezultat__

Veƒái koeficijent sreƒáe ne znaƒçi da ƒáe dr≈æava da ima manju stopu samoubistva.

__Uticaj konzumacije alkohola na stopu samoubistva__

```
#sortiramo podatke po stopi samoubistva opadajuƒáe
df = newAlcDF.sort(newAlcDF.suicide_rate.desc())
df = df.where((df.both != 0) & (df.male != 0) & (df.female != 0))
df.display()
df.createOrReplaceTempView("df_table")
```

Na gornjem primeru mo≈æemo videti da u nekim sluƒçajevima sa veƒáom konzumpcijom alkohola raste i stopa samoubistva

```
%sql
SELECT country from df_table where female > male

```

Mo≈æemo primetiti da ne postoji dr≈æava gde ≈æene imaju veƒáu konzumpciju alkohola od mu≈°karca

```
display(df.take(15))
display(df.tail(15))
```

Na gornjim primerima mo≈æemo videti da u veƒáini primera sa manjom konzumpcijom alkohola imamo i manju stopu samoubistva

```
#Sortiramo podatke po konzumpciji alkohola opadajuƒáe
df = df.sort(df.both.desc())
display(df.take(15))
display(df.tail(15))
```

Kao i na prethodnom primeru mo≈æemo videti da u veƒáini sluƒçajeva manja konzumpcija alkohola znaƒçi i manja stopa samoubistva

__Rezultat__

Konzumpcija alkohola mo≈æe da utiƒçe na stopu samoubistva u dr≈æavi ali ne u prevelikoj meri.
