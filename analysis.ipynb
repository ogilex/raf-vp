```
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```
#Učitamo transformisane podatke

file_path = "/tmp/happiness/"

hapDF = spark.read.csv(file_path, header=True, inferSchema=True)

hapDF.display()
```

__Prikaz mape sveta sa koeficijentom sreće__

```
newHapDF = hapDF.join(suDF, hapDF.country == suDF.country, "inner").drop(suDF.country).drop("year")
display(newHapDF)
```

```
#Učitamo transformisane podatke
file_path = "/tmp/alcohol/"

alcDF = spark.read.csv(file_path, header=True, inferSchema=True)
alcDF.display()
```

__Prikaz mape sveta sa konzumpcijom alkohola za oba pola__

```
newAlcDF = alcDF.join(suDF, alcDF.country == suDF.country, "inner").drop(suDF.country).drop("year")
display(newAlcDF)
```

__Prikaz mape sveta sa stopom samoubistva__

```
#Učitamo transformisane podatke
file_path = "/tmp/suicide/"

suDF = spark.read.csv(file_path, header=True, inferSchema=True)
suDF.display() 
```

__Uticaj alkohola na sreću naroda u državi__

```
joinedDF = newHapDF.join(newAlcDF, newHapDF.country == newAlcDF.country, "inner").drop(newAlcDF.country).drop(newAlcDF.gdp).select("rank","country","score","both","male","female", "gdp")
joinedDF.display()
df = joinedDF.sort(joinedDF.score.desc())
```

```

df.createOrReplaceTempView("df_table")
print(df.head())
print(df.tail(1))
```

Iz gornjih podataka možemo videti da Finska ima najveci koeficijent sreće 7.769, a Centralnoafrička republika ima najmanji koeficijent sreće 3.083. Takođe možemo primetiti da Finska konzumira 10.65 litra čistog alkohola 

godišnje, što je znazno veće od Centralnoafričke republike koja konzumira 1.71 litra čistog alkogola godišnje. Ovo može da znači da je Finska srećnija jer konzumira više alkohola, ali može biti i do razlike u ekonomskoj

razvijenosti zemlje. Za sledeći primer uzećemo zemlje sa sličnim ekonomskim stanjem.

```
%sql
SELECT country, score, both, gdp from df_table where country == "france" OR country == "italy"
```

U gornjem primeru vidimo da Francuska konzumira više alkohola i da je koeficijent sreće veći u odnosu na Italiju

```
%sql
SELECT country, score, both, gdp from df_table where country == "germany" OR country == "austria"
```

U gornjem primeru vidimo da iako Nemačka konzumira više alkohola, nema veći koeficijent sreće. Ovo može da znaći da alkohol nema uticaj na koeficijent sreće države, ali isto moramo uzeti u obzir da je nacionalno piće Nemačke pilsner što je pivo, a nacionalno piće Austrije je almdudler što je tip soka.

```
%sql
SELECT country, score, both, gdp from df_table where country == "latvia" OR country == "turkey"
```

Za poslednji primer ćemo uzeti Latviu i Tursku. Latvija je zemlja sa najvišom konzumpcijom alkohola, a Turska je zemlja sa veoma niskom konzumpcijom alkohola. Razlika u konzumpciji alkohola je ogromna u odnosu na razliku u 

koeficijentu sreće.

__Rezultat__

U većinu primera vidimo korelaciju izmedju konzumpcije alkohola i većeg koeficijenta sreće u državi.

__Da li srećnije zemlje nužno imaju manju stopu samoubistava__

```
joinedDF = hapDF.join(suDF, hapDF.country == suDF.country, "inner").drop(suDF.country).select("rank", "country","score","suicide_rate", "gdp")
display(joinedDF)
```

U gornjem primeru možemo da vidimo da zemlje sa višim koeficijentom sreće mogu imati skoro duplo veću stopu samoubistva od zemlja sa nižim koeficijentom sreće

```
joinedDF = joinedDF.sort(joinedDF.score.desc())
df = joinedDF
df.createOrReplaceTempView("df_table")
```

```
display(df.take(10))
display(df.tail(10))

```

U gornjem primeru imamo 10 zemlja sa najvećim koeficijentom sreće i 10 zemlja sa najnižim koeficijentoim sreće. Možemo primetiti da FInska koja ima najveći koeficijent sreće ima skoro 4 puta veću stopu samoubistva od Afganistana koji ima duplo manji koeficijent sreće

```
df = df.sort(df.suicide_rate.desc())
display(df.take(10))
display(df.tail(10))
```

U gornjem primeru imamo 10 zemlja sa najvećom stopom samoubistva i 10 zemlja sa najnižom stopom samoubistva. U prvom grafu se većinski nalaze zemlje kategarizovane u zemlje drugog sveta, pored Lesata koji je

zemlja trećeg sveta i Slovenije koja može da se smatra zemljom prvog sveta. U donjem grafu su sve zemlje trećeg sveta. Zdravorazumski zar ne bi trebalo da razvijenije zemlje imaju veći koeficijent sreće i manju stopu samoubistva

ali ovde vidimo da to nije uvek istina.

__Rezultat__

Veći koeficijent sreće ne znači da će država da ima manju stopu samoubistva.

__Uticaj konzumacije alkohola na stopu samoubistva__

```
#sortiramo podatke po stopi samoubistva opadajuće
df = newAlcDF.sort(newAlcDF.suicide_rate.desc())
df = df.where((df.both != 0) & (df.male != 0) & (df.female != 0))
df.display()
df.createOrReplaceTempView("df_table")
```

Na gornjem primeru možemo videti da u nekim slučajevima sa većom konzumpcijom alkohola raste i stopa samoubistva

```
%sql
SELECT country from df_table where female > male

```

Možemo primetiti da ne postoji država gde žene imaju veću konzumpciju alkohola od muškarca

```
display(df.take(15))
display(df.tail(15))
```

Na gornjim primerima možemo videti da u većini primera sa manjom konzumpcijom alkohola imamo i manju stopu samoubistva

```
#Sortiramo podatke po konzumpciji alkohola opadajuće
df = df.sort(df.both.desc())
display(df.take(15))
display(df.tail(15))
```

Kao i na prethodnom primeru možemo videti da u većini slučajeva manja konzumpcija alkohola znači i manja stopa samoubistva

__Rezultat__

Konzumpcija alkohola može da utiče na stopu samoubistva u državi ali ne u prevelikoj meri.

__Zaključak__

Ako uzmemo rezultate hipoteze iz ove analize dobićemo da države koje imaju veću konzumpciju alkohola su srećnije, da zemlje koje imaju veći koeficijent sreće nemaju nužno manju stopu stopu samoubistva, ponekad imaju i veću od država koje imaju manji koeficijent sreće i da konzumpcija alkohola povećava stopu samoubistva. Ovo znači da ovaj način merenja sreće nije adekvatan ili da srećni ljudi izvršavaju samoubistvo i da više konzumiraju alkohol.
